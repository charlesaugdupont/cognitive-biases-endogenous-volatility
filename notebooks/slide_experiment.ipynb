```python
import sys
sys.path.append("../")

from utils import *
from experiment import unpack_and_dequantize
import pickle
import os
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({"font.size" : 15,
                     "figure.dpi" : 100, 
                     "grid.alpha" : 0.3, 
                     "axes.grid": False, 
                     "axes.axisbelow" : True,
                     "figure.figsize":(8,6),
                     "mathtext.fontset":"cm",
                     "xtick.labelsize": 14,
                     "ytick.labelsize": 14,
                     "axes.labelsize": 16, 
                     "legend.fontsize": 13.5})
plt.rc("text", usetex=False)
plt.rc("font", family="serif")
```

# $\gamma$

```python
D = "slide_gamma"
IDX = 1
ordered_files = sorted(os.listdir("../"+D), key=lambda x: float(x.split("_")[IDX]))

fig, axs = plt.subplots(5, 5, figsize=(12,12), sharex=True, sharey=True)
for idx, file in enumerate(ordered_files):
    with open(os.path.join("../"+D,file), "rb") as f:
        res = pickle.load(f)
    gamma = float(file.split("_")[IDX])
    axs[int(idx/5)][idx%5].scatter(
        unpack_and_dequantize(res["wealth"][:,-1], 200),
        unpack_and_dequantize(res["health"][:,-1], 200),
        s=1,
        alpha=0.7,
        color="dodgerblue"
    )
    axs[int(idx/5)][idx%5].imshow(res["policy"].T, origin="lower")
    axs[int(idx/5)][idx%5].set_xlim(0,201)
    axs[int(idx/5)][idx%5].set_ylim(0,201)
    axs[int(idx/5)][idx%5].set_title(fr"$\gamma={gamma:.5f}$", fontsize=12)

fig.tight_layout()
plt.show()
```
