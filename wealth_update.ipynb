```python
def wealth_growth_rate(u, N, Gmin=-0.05, Gmax=0.05):
    u_norm = (u - 1) / (N - 1)
    return 0.05 * (2*u_norm - 1)
```

```python
util = np.linspace(1,200,1000)

plt.plot(util, wealth_growth_rate(util, N=200))
plt.yticks(np.linspace(-0.05, 0.05, 5))
plt.show()
```
