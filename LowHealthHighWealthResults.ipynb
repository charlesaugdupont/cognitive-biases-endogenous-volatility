### Sen's Welfare Function

```python
def gini(data):
    if len(data) == 0:
        return 0
    data = np.asarray(data)
    data = np.sort(data)
    n = len(data)
    index = np.arange(1, n + 1)  # 1-based index
    gini = (2 * np.sum(index * data)) / (n * np.sum(data)) - (n + 1) / n
    return gini

def sen_welfare(data):
    return np.mean(data) * (1-gini(data))
```

### Limit Cycle Detection

```python
MIN_VAL, MAX_VAL, CELL_SIZE = 1, 200, 30
NUM_CELLS_X = (MAX_VAL - MIN_VAL) // CELL_SIZE + 1

def periodic_likelihood(
    wealth, 
    health, 
    min_range_threshold_factor=0.5,
    spectral_purity_threshold=0.75,
    min_power_threshold=1.0
):
    """
    The most robust method for periodicity detection, incorporating a
    detrending step to prevent confusion between trends and oscillations.

    Args:
        wealth (np.array): Array of wealth values.
        health (np.array): Array of health values.
        min_range_threshold_factor (float): Minimum movement range required,
            as a factor of CELL_SIZE.
        spectral_purity_threshold (float): Minimum ratio of the dominant
            frequency's power to the total signal power.
        min_power_threshold (float): The minimum total power the detrended
            signal must have to be considered for periodicity. This rejects
            signals that are just flat lines or noise after detrending.
    """
    # 1. State-Space Extent Check (as before)
    # -----------------------------------------------------------------
    wealth_range = np.ptp(wealth)
    health_range = np.ptp(health)
    min_movement_range = CELL_SIZE * min_range_threshold_factor
    if wealth_range < min_movement_range and health_range < min_movement_range:
        return False

    # 2. Discretization and Signal Creation (as before)
    # -----------------------------------------------------------------
    grid_indices = ((np.stack([wealth, health], axis=-1) - MIN_VAL) // CELL_SIZE).astype(np.int32)
    i, j = grid_indices[..., 1], grid_indices[..., 0]
    signal = i * NUM_CELLS_X + j
    
    # 3. --- NEW: Detrend the Signal ---
    # -----------------------------------------------------------------
    # This removes the linear trend, which was causing the false positive.
    detrended_signal = sp_signal.detrend(signal)
    
    # 4. Spectral Purity Analysis on the DETRENDED signal
    # -----------------------------------------------------------------
    n = len(detrended_signal)
    if n < 2: return False

    # Compute the one-sided Power Spectral Density (PSD)
    fft_coeffs = np.fft.rfft(detrended_signal)
    psd = np.abs(fft_coeffs)**2
    
    # Calculate total power (the DC component is already gone due to detrending)
    total_power = np.sum(psd)
    
    # --- NEW: Power Check ---
    # If the total power is negligible after detrending, it was just a trend.
    if total_power < min_power_threshold:
        return False
        
    # Find the power of the dominant frequency peak
    dominant_peak_power = np.max(psd)
    
    # Calculate spectral purity
    spectral_purity = dominant_peak_power / total_power
    
    # 5. Final Decision (as before)
    # -----------------------------------------------------------------
    return spectral_purity
```

```python
def get_rotational_consistency(trajectory_segment, smoothing_window=5):
    """
    Analyzes a trajectory to determine if its movement is consistently rotational (a cycle)
    or random by using a more robust method based on smoothed velocity.

    Returns:
        A float between 0.5 and 1.0. 
        - A value near 1.0 means highly consistent rotation.
        - A value near 0.5 means random, non-directional movement.
    """
    if len(trajectory_segment) < 15: # Need enough points for smoothing and analysis
        return 0.5

    # 1. Smooth the trajectory to remove high-frequency noise
    # Using a simple moving average (convolution)
    window = np.ones(smoothing_window) / smoothing_window
    # Use 'valid' mode to avoid edge effects; this shortens the array
    smoothed_x = np.convolve(trajectory_segment[:, 0], window, mode='valid')
    smoothed_y = np.convolve(trajectory_segment[:, 1], window, mode='valid')
    smoothed_traj = np.column_stack([smoothed_x, smoothed_y])
    
    if len(smoothed_traj) < 2:
        return 0.5

    # 2. Calculate the centroid of the SMOOTHED trajectory
    centroid = np.mean(smoothed_traj, axis=0)
    centered_traj = smoothed_traj - centroid

    # 3. Calculate velocity vectors for the smoothed trajectory
    # v(t) = r(t) - r(t-1)
    velocities = np.diff(smoothed_traj, axis=0)
    
    # We must align the centered position vectors with the velocities
    # velocities[i] corresponds to the transition from centered_traj[i] to centered_traj[i+1]
    # So we use centered_traj[:-1]
    positions_for_cross = centered_traj[:-1]
    
    if len(positions_for_cross) == 0:
        return 0.5

    # 4. Calculate the 2D cross-product of position and velocity (r x v)
    cross_products = np.cross(positions_for_cross, velocities)
    
    # Ignore steps with no effective rotation
    cross_products = cross_products[np.abs(cross_products) > 1e-6]

    if len(cross_products) == 0:
        return 0.5 # No rotational movement detected

    # 5. Count positive and negative rotations
    num_positive = np.sum(cross_products > 0)
    num_negative = np.sum(cross_products < 0)

    # 6. Consistency is the ratio of the dominant direction to the total
    consistency = max(num_positive, num_negative) / len(cross_products)
    
    return consistency

def get_attractors(trajectories):
    # --- Algorithm Parameters (TUNE THESE FOR YOUR DATA) ---
    N_STEPS_TO_CONSIDER = 5  # How many of the last time steps to analyze
    DBSCAN_EPS = 8.0          # DBSCAN radius for finding attractors
    DBSCAN_MIN_SAMPLES = 5    # Min points to form a dense region
    POINT_VAR_THRESHOLD = 2.0 # Max variance for a cluster to be a point
    CHARACTERIZATION_WINDOW = 1000
    MIN_CYCLE_DISPLACEMENT = 10
    CYCLE_CONSISTENCY_THRESHOLD = 0.70
    SMOOTHING_WINDOW_SIZE = 20

    point_cloud = []
    agent_id_map = []
    segment_id_map = []
    segment_repo = {}
    
    segment_id_counter = 0
    for agent_id, traj in trajectories.items():
        if len(traj) >= N_STEPS_TO_CONSIDER:
            last_n_points = traj[-N_STEPS_TO_CONSIDER:]
            point_cloud.extend(last_n_points)
            agent_id_map.extend([agent_id] * N_STEPS_TO_CONSIDER)
            segment_id_map.extend([segment_id_counter] * N_STEPS_TO_CONSIDER)
            segment_repo[segment_id_counter] = last_n_points
            segment_id_counter += 1
            
    point_cloud = np.array(point_cloud)
    agent_id_map = np.array(agent_id_map)
    segment_id_map = np.array(segment_id_map)
    
    # --- Step 2: Perform DBSCAN Clustering ---
    db = DBSCAN(eps=DBSCAN_EPS, min_samples=DBSCAN_MIN_SAMPLES).fit(point_cloud)
    cluster_labels = db.labels_
    
    unique_labels = set(cluster_labels)
    n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
    n_noise = np.sum(cluster_labels == -1)
    
    attractor_summary = {}
    for label in unique_labels:
        if label == -1: continue
    
        cluster_mask = (cluster_labels == label)
        cluster_points = point_cloud[cluster_mask]
        
        # Get all agents that contributed to this cluster
        agents_in_cluster = set(agent_id_map[cluster_mask])
    
        # --- START OF MODIFIED LOGIC ---
    
        # A) Check for Point Attractor (no change)
        cluster_variance = np.var(cluster_points[:, 0]) + np.var(cluster_points[:, 1])
        if cluster_variance < POINT_VAR_THRESHOLD:
            attractor_type = "Point Attractor"
        else:
            # B) Select a representative agent from the cluster
            representative_agent_id = list(agents_in_cluster)[0]
            
            # C) Get its LONG historical trajectory for analysis
            # Using a longer window here just for characterization
            long_trajectory_segment = trajectories[representative_agent_id][-CHARACTERIZATION_WINDOW:]
    
            # D) Check for Transient Path using the long trajectory
            start_point = long_trajectory_segment[0]
            end_point = long_trajectory_segment[-1]
            
            # Use a threshold relative to the overall size of the trajectory
            path_length = np.sum(np.linalg.norm(np.diff(long_trajectory_segment, axis=0), axis=1))
            endpoint_distance = np.linalg.norm(start_point - end_point)
    
            if endpoint_distance / path_length > 0.5: # If endpoint is far from start relative to path length
                attractor_type = "Slow Manifold"
            else:
                attractor_type = "Stochastic Attractor"
                consistency = get_rotational_consistency(long_trajectory_segment)
                if consistency > CYCLE_CONSISTENCY_THRESHOLD:
                    x_range = np.ptp(long_trajectory_segment[:, 0])
                    y_range = np.ptp(long_trajectory_segment[:, 1])
                    if x_range >= MIN_CYCLE_DISPLACEMENT and y_range >= MIN_CYCLE_DISPLACEMENT:
                        attractor_type = "Cycle"
    
        attractor_summary[label] = {
            "type": attractor_type,
            "num_agents": len(agents_in_cluster),
            "centroid": np.mean(cluster_points, axis=0)
        }

    return attractor_summary
```

```python
def compute_flux(wealth, health, a, b, delta=4):
    inn, out = 0, 0
    visits = np.zeros(wealth.shape[0])
    for t in range(wealth.shape[1]):        
        # IN
        if t == 0:
            reached = np.where( (np.abs(wealth[:,t] - a) <= delta) & (np.abs(health[:,t] - b) <= delta) )[0]
        else:
            reached = np.where( ((np.abs(wealth[:,t] - a) <= delta) & (np.abs(health[:,t] - b) <= delta)) & 
                                 ((np.abs(wealth[:,t-1] - a) > delta) | (np.abs(health[:,t-1] - b) > delta)) )[0]
        inn += len(reached)
        visits[reached] +=1
        # OUT
        if t > 0:
            left = np.where( ((np.abs(wealth[:,t] - a) > delta) | (np.abs(health[:,t] - b) > delta)) & 
                             ((np.abs(wealth[:,t-1] - a) <= delta) & (np.abs(health[:,t-1] - b) <= delta)) )[0]
            out += len(left)

    return (inn,out), visits


def is_periodic(signal):
    """
    Determine if a signal is periodic using autocorrelation.

    Parameters:
    signal (array-like): The input signal.
    threshold (float): The threshold for detecting periodicity. Default is 0.8.

    Returns:
    bool: True if the signal is periodic, False otherwise.
    """
    autocorr = correlate(signal, signal, mode='full')
    autocorr = autocorr[autocorr.size // 2:]
    max_autocorr = np.max(autocorr)
    second_max_autocorr = np.sort(autocorr)[-2]
    return (second_max_autocorr / max_autocorr)


def radar_plot(X1, X2, labels):
    colors = ['dodgerblue', 'crimson']
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    
    for i,arr in enumerate([X1, X2]):
        means = np.mean(arr, axis=0)
        stds = np.std(arr, axis=0)
        
        num_vars = len(means)
        
        angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
        angles += angles[:1]  # Close the polygon
        
        values = means.tolist()
        values += values[:1]
        
        std_upper = (means + stds).tolist()
        std_upper += std_upper[:1]
        std_lower = (means - stds).tolist()
        std_lower += std_lower[:1]
        
        ax.plot(angles, values, color=colors[i], linewidth=1, label=labels[i])
        ax.fill_between(angles, std_lower, std_upper, color=colors[i], alpha=0.1)

    labels = [r"$\alpha$", r"$P_{H_-}$", r"$P_{H_+}$", r"$\gamma$", r"$A$", r"$\omega$", r"$\eta$"]
    ax.legend()
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    plt.show()
```

```python
flux_result = []
for idx,f in tqdm(enumerate(os.listdir(result_dir))):
    with open(os.path.join(result_dir, f), "rb") as f:
        res = pickle.load(f)
    wealth = res["wealth"]
    health = res["health"]
    r = {}
    for m in MINIMA[idx][1]:
        a,b = m[0]
        in_out_counts, visits = compute_flux(wealth, health, a, b)
        r[(a,b)] = (in_out_counts, visits)
    flux_result.append((idx, r))

with open(f"alternate_flux_{result_suffix}.pickle", "wb") as f:
    pickle.dump(flux_result, f)
```

```python
periodic_looping = {}

for idx, f in tqdm(enumerate(os.listdir(result_dir))):

    with open(os.path.join(result_dir, f), "rb") as file:
        res = pickle.load(file)
    wealth = res["wealth"]
    health = res["health"]

    mins = MINIMA[idx][1]
    if len(mins) == 1:
        periodic_looping[idx] = periodic_count = np.zeros((10000,))
        continue
    min_dict = {
        (mins[i][0][0], mins[i][0][1]): i for i in range(len(mins))
    }
    
    minima_array = np.array(list(min_dict.keys()))
    
    # Stack wealth and health into shape (10000, 5000, 2)
    agent_positions = np.stack((wealth, health), axis=-1)

    # Compute distances to each minimum
    distances = np.linalg.norm(
        agent_positions[:, :, None, :] - minima_array[None, None, :, :], axis=-1
    )

    # Get the index of the closest minimum
    closest_minima = np.argmin(distances, axis=-1)

    periodic_count = np.zeros((10000,))
    for j,row in enumerate(closest_minima):
        change_indices = np.where(np.diff(row) != 0)[0] + 1
        collapsed = row[np.insert(change_indices, 0, 0)]
        if len(collapsed) < 5:
            continue
        periodic_count[j] = is_periodic(collapsed[2:])

    periodic_looping[idx] = periodic_count

with open(f"alternate_periodic_looping_{result_suffix}.pickle", "wb") as f:
    pickle.dump(periodic_looping, f)
```
